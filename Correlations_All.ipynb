```python
plt.figure(figsize=(35,35))
sns.set(style='dark')
sns.heatmap(df.corr(), center=0, cmap='YlGnBu', annot=True, fmt='.2f')
plt.savefig("Correlation_matrix_with_the_original_data.pdf")
correlation_matrix = df.corr()
```

```python
correlation_matrix_filtered = correlation_matrix[correlation_matrix < 0]
plt.figure(figsize=(35, 35))
sns.set(style='dark')
sns.heatmap(correlation_matrix_filtered, center=0, cmap='YlGnBu', annot=True, fmt='.2f')
plt.savefig("Correlation_matrix_with_the_original_data_lower_0_percent.pdf")
```
